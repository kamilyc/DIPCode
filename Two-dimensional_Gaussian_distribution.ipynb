# 二维高斯分布特性观察

```python
# 导入必要库
import matplotlib.pyplot as plt
import cv2
import numpy as np
from scipy import interpolate
%matplotlib inline
np.set_printoptions(suppress=True, threshold=np.nan)
```

## 定义高斯函数

```python
def guassXY(x,y,sig1,sig2,rou = 0.9):
    return (1/(2*np.pi*sig1*sig2*2*np.sqrt(1-rou*rou)))*np.exp((-1/(2*(1-rou*rou)))*(x*x/(sig1*sig1) - rou*x*y/2 + y*y/(sig2*sig2)))

```

```python
#产生需要的坐标
x = np.array(np.arange(-3,3,0.1))
y = np.array(np.arange(-3,3,0.1))
xv, yv = np.meshgrid(x, y)
```

## 协方差从-0.9到0.9变化时二维高斯分布变化

```python
rst1 = guassXY(xv,yv,1,2,-0.9)
rst2 = guassXY(xv,yv,1,2,-0.4)
rst3 = guassXY(xv,yv,1,2,0)
rst4 = guassXY(xv,yv,1,2,0.4)
rst5 = guassXY(xv,yv,1,2,0.9)

plt.subplot(231),plt.imshow(rst1,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(rst2,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(rst3,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(rst4,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(rst5,cmap = 'gray')
plt.xticks([]), plt.yticks([])
```

## 方差对调时高斯分布变化

```python
rst6 = guassXY(xv,yv,1,2,0.9)
plt.imshow(rst6,cmap = 'gray'),plt.title('sigma1=1,sigma2=2,rou=0.9'),plt.axis("off"),plt.show()

rst7 = guassXY(xv,yv,2,1,0.9)
plt.imshow(rst7,cmap = 'gray'),plt.title('sigma1=2,sigma2=1,rou=0.9'),plt.axis("off"),plt.show()
```

## 观察协方差变化时傅里叶变换的变化

```python
f1 = np.fft.fft2(rst1)
fshift1 = np.fft.fftshift(f1)
magnitude_spectrum1 = np.log(np.abs(fshift1))

f2 = np.fft.fft2(rst2)
fshift2 = np.fft.fftshift(f2)
magnitude_spectrum2 = np.log(np.abs(fshift2))

f3 = np.fft.fft2(rst3)
fshift3 = np.fft.fftshift(f3)
magnitude_spectrum3 = np.log(np.abs(fshift3))

f4 = np.fft.fft2(rst4)
fshift4 = np.fft.fftshift(f4)
magnitude_spectrum4 = np.log(np.abs(fshift4))

f5 = np.fft.fft2(rst5)
fshift5 = np.fft.fftshift(f5)
magnitude_spectrum5 = np.log(np.abs(fshift5))

plt.subplot(231),plt.imshow(magnitude_spectrum1,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(magnitude_spectrum2,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(magnitude_spectrum3,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(magnitude_spectrum4,cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(magnitude_spectrum5,cmap = 'gray')
plt.xticks([]), plt.yticks([])
```
